```python
import pandas as pd
def remove_bmi_hi_rows(filename):
  df = pd.read_csv(filename)
  df.drop(df[df['bmi'] == 'NA'].index, inplace=True)
  df.to_csv("/content/drive/MyDrive/Colab Notebooks/NAFLD/nafld_clean.csv", index=False)
filename = "/content/drive/MyDrive/Colab Notebooks/NAFLD/nafld.csv"
remove_bmi_hi_rows(filename)
print(f"Rows with bmi='hi' removed from '{filename}'.")
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/NAFLD/nafld_clean.csv')
df = df.dropna(subset=['bmi'])
df.to_csv('/content/drive/MyDrive/Colab Notebooks/NAFLD/nafld_clean.csv', index=False)
```
